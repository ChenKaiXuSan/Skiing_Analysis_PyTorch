```python
import sys
sys.path.append("/workspace/code")
from vis_3d_kpt.visualize import  setup_visualizer
```

```python
skeleton, scene = setup_visualizer()
```

```python
# 确定不同视角的方位角和仰角
from matplotlib import pyplot as plt

for azim in range(-180, 180, 90):
    for elev in range(0, 180, 90):
        # y to -z, z to y
        pose_3d_kpts = fused_3d_kpts.copy()
        pose_3d_kpts[:, [1, 2]] = pose_3d_kpts[:, [2, 1]] # swap y and z
        pose_3d_kpts[:, 2] = -pose_3d_kpts[:, 2] # ivert z
        
        fig = scene.draw_scene(ax=None, kpts_world=pose_3d_kpts, elev=elev, azim=azim)
        print(f"azim: {azim}, elev: {elev}")
        
        plt.show()
```
